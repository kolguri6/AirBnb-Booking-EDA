#### **Chart - 9 Average month reviews**

````python
# Chart - 9 visualization code
# Convert the 'last_review' column to a datetime object
Airbnb_df['last_review_month'] = pd.to_datetime(Airbnb_df['last_review']).dt.month

reviews_per_month = Airbnb_df.groupby(['last_review_month'])['number_of_reviews'].mean() #Groups all unique values in column
reviews_per_month = pd.DataFrame(reviews_per_month).reset_index() #converts into dataframe
reviews_per_month = reviews_per_month.rename(columns = {'last_review_month': 'month', 'number_of_reviews':'AVG_reviews'}) # renaming the columns
reviews_per_month
````
