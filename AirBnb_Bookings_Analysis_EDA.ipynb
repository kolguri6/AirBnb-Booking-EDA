#### Chart - 9 Average month reviews

````python
# Chart - 9 visualization code
# Convert the 'last_review' column to a datetime object
Airbnb_df['last_review_month'] = pd.to_datetime(Airbnb_df['last_review']).dt.month

reviews_per_month = Airbnb_df.groupby(['last_review_month'])['number_of_reviews'].mean()
reviews_per_month = pd.DataFrame(reviews_per_month).reset_index()
reviews_per_month = reviews_per_month.rename(columns = {'last_review_month': 'month', 'number_of_reviews':'AVG_reviews'})
reviews_per_month
````

````python
plt.plot(reviews_per_month['month'], reviews_per_month['AVG_reviews'])
plt.xlabel('Month')
plt.ylabel('Average Number of Reviews')
plt.title('Average Number of Reviews Per Month')
plt.grid(True)
plt.show()
````
